```python
# Create a Year column from the Reported Date. 
# This will help us query the data for the year we are interested in. 
Denver_Crimes['year'] = pd.DatetimeIndex(Denver_Crimes['REPORTED_DATE']).year
Denver_Crimes.head()
```

```python
# Remove columns you wont need. 
# OFFENSE_ID, OFFENSE_CODE 
# Denver_Crimes = Denver_Crimes.drop('OFFENSE_ID', 1) #If we ran this twice, we would get an error
# Denver_Crimes.drop(['OFFENSE_CODE', 'OFFENSE_CODE_EXTENSION'], axis=1, inplace=True)
# Denver_Crimes.shape
Denver_Crimes.head()
```

```python
# Move the year Column 
# This code moves the column year to the front. 
Denver_Crimes = Denver_Crimes[['year'] + [ col for col in Denver_Crimes.columns if col != 'year']]
Denver_Crimes.head()
```

```python
# Create a new data frame to seperate out the data. 
# Query the year column for all crimes in 2020.
Denver_Crimes_2020 = Denver_Crimes[Denver_Crimes.year == 2020]
Denver_Crimes_2020.head()
```

```python
# Convert Reported Date to Date type to remove the time stamps that you dont need. 
# this will be helpful when we create our heat map calendar. 
Denver_Crimes_2020['timestamp'] = pd.to_datetime(Denver_Crimes_2020['REPORTED_DATE'])
Denver_Crimes_2020['New_Date'] = Denver_Crimes_2020['timestamp'].dt.date
Denver_Crimes_2020['Months'] =  pd.to_datetime(Denver_Crimes_2020['New_Date']).dt.month
Denver_Crimes_2020.head()
```

```python
Denver_Crimes_2020['monthName'] = Denver_Crimes_2020['Months'].apply(lambda x: calendar.month_name[x])
Denver_Crimes_2020
```

```python
# Remove the timestamp column
Denver_Crimes_2020 = Denver_Crimes_2020.drop(columns=['timestamp'])

# Get new size of new data frame 
Denver_Crimes_2020.shape
```

```python
#Reindex the dataframe 
Denver_Crimes_2020 = Denver_Crimes_2020.reset_index(drop=True)
Denver_Crimes_2020.head()
```
